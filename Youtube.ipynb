### Para baixar playlist de vídeos

```python
from pytube import YouTube, Playlist
from time import sleep

PLAYLIST_URL = 'https://youtube.com/playlist?list=PLfCKf0-awunPFkWOKWNaXB_ndaHBlJ4QQ&si=3htst2Tk7NOw9baZ'
playlist = Playlist(PLAYLIST_URL)

for url in playlist[367:367]:
    video = YouTube(url)
    stream = video.streams.get_highest_resolution()
    stream.download(output_path='playlist')
    sleep(2)
```

### Verificar quantos vídeos existem na playlist

```python
tot = 0
for url in playlist:
    print(url)
    tot += 1
print('-' * 30)
print(f'Existe um total de {tot} vídeos nesta playlist.')
```

### Para baixar uma playlist de áudio

```python
from pytube import YouTube, Playlist

PLAYLIST_URL = 'https://youtube.com/playlist?list=PLeOj4bdrCSC81-ffLj74Q0Ie7KIvxT1hU&si=u6P2jji9O2Lm4YkL'
playlist = Playlist(PLAYLIST_URL)

for url in playlist:
    video = YouTube(url)
    
    # Obter todas as streams de áudio disponíveis
    audio_streams = video.streams.filter(only_audio=True)
    
    if audio_streams:
        # Baixar a primeira stream de áudio disponível
        audio = audio_streams[0]
        audio.download()
    else:
        print(f'Não há streams de áudio disponíveis para o vídeo: {url}')
```
